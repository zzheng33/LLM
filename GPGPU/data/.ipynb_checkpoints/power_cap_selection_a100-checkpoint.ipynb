# EDP

```python
import ast
import pandas as pd
import os
from itertools import product

performance_matrix_path = "./prediction_res/performance_matrix.csv"
power_cap_res_path = "./ecp_power_cap_res/2_dual_cap/"

# Applications of interest
apps = ['miniGAN', 'sw4lite', 'bert_large', 'UNet', 'Resnet50', 'lammps', 'gromacs', 'NAMD']
# apps = ['bert_large']

# Load performance matrix
performance_matrix = pd.read_csv(performance_matrix_path, index_col=0)

# Dictionary to store results
edp_results = {}

cpu_caps = [200, 190, 180, 170, 160, 150, 140]
gpu_caps = [250, 240, 230, 220, 210, 200, 190, 180, 170, 160, 150]
power_cap_pairs = list(product(cpu_caps, gpu_caps))

for app in apps:
    app_path = os.path.join(power_cap_res_path, app)
    

    min_edp = float('inf')
    best_power_pair = None

    for power_pair in power_cap_pairs:
        cpu_power, gpu_power = power_pair  # Unpack tuple

        # Get normalized performance
        power_pair = f"({cpu_power}, {gpu_power})"

        norm_performance = performance_matrix.loc[power_pair, app]
        # print(norm_performance)

        # Compute normalized execution time
        norm_exec_time = 1 / norm_performance

        # Load CPU and GPU power files
        cpu_power_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_cpu_power.csv")
        gpu_power_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_gpu_metrics.csv")

   

        cpu_data = pd.read_csv(cpu_power_file)
        gpu_data = pd.read_csv(gpu_power_file)

        # Compute CPU energy consumption
        cpu_energy = (cpu_data["Package Power (W)"].mean() + cpu_data["DRAM Power (W)"].mean()) * norm_exec_time

        # Compute GPU energy consumption
        gpu_energy = gpu_data["Power (W)"].mean() * norm_exec_time

        
        # Total energy consumption
        total_energy = cpu_energy + gpu_energy
        

        # Compute EDP
        edp = total_energy * norm_exec_time

        # Update best power pair
        if edp < min_edp:
            min_edp = edp
            best_power_pair = (cpu_power, gpu_power)

        

    # Step 3: Compute EDP for actual performance using the selected power cap pair and baseline (200_250)
    if best_power_pair:
        cpu_power, gpu_power = best_power_pair

        # Baseline power cap (200_250)
        baseline_cpu_file = os.path.join(app_path, "200_250_cpu_power.csv")
        baseline_gpu_file = os.path.join(app_path, "200_250_gpu_metrics.csv")

    
        baseline_cpu_data = pd.read_csv(baseline_cpu_file)
        baseline_gpu_data = pd.read_csv(baseline_gpu_file)

        # Execution time from the last row of Time (s)
        baseline_exec_time = baseline_cpu_data["Time (s)"].iloc[-1]

        # Compute CPU energy for baseline
        baseline_cpu_energy = (baseline_cpu_data["Package Power (W)"].mean() + 
                               baseline_cpu_data["DRAM Power (W)"].mean()) * baseline_exec_time

        # Compute GPU energy for baseline
        baseline_gpu_energy = baseline_gpu_data["Power (W)"].mean() * baseline_exec_time

        # Total energy for baseline
        baseline_total_energy = baseline_cpu_energy + baseline_gpu_energy

        # Compute baseline EDP
        baseline_edp = baseline_total_energy * baseline_exec_time

        

        # Compute EDP for selected power cap pair
        selected_cpu_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_cpu_power.csv")
        selected_gpu_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_gpu_metrics.csv")

        if os.path.exists(selected_cpu_file) and os.path.exists(selected_gpu_file):
            selected_cpu_data = pd.read_csv(selected_cpu_file)
            selected_gpu_data = pd.read_csv(selected_gpu_file)

            # Execution time from the last row of Time (s)
            selected_exec_time = selected_cpu_data["Time (s)"].iloc[-1]

            # Compute CPU energy for selected power cap
            selected_cpu_energy = (selected_cpu_data["Package Power (W)"].mean() + 
                                   selected_cpu_data["DRAM Power (W)"].mean()) * selected_exec_time

            # Compute GPU energy for selected power cap
            selected_gpu_energy = selected_gpu_data["Power (W)"].mean() * selected_exec_time

            # Total energy for selected power cap
            selected_total_energy = selected_cpu_energy + selected_gpu_energy

            # Compute EDP for selected power cap
            selected_edp = selected_total_energy * selected_exec_time

            # Store results
            edp_results[app] = {
                "Best Power Pair": f"{cpu_power}_{gpu_power}",
                "Baseline EDP": baseline_edp,
                "Selected EDP": selected_edp
            }
```

```python
# Calculate EDP savings for each application
for app, results in edp_results.items():
    baseline_edp = results["Baseline EDP"]
    selected_edp = results["Selected EDP"]
    
    # Compute EDP saving percentage
    edp_saving = ((baseline_edp - selected_edp) / baseline_edp) * 100

    # Print results with EDP savings
    print(f"Application: {app}")
    print(f"Best Power Pair: {results['Best Power Pair']}")
    print(f"Baseline EDP: {baseline_edp:.4f}")
    print(f"Selected EDP: {selected_edp:.4f}")
    print(f"EDP Saving: {edp_saving:.2f}%")
    print("-" * 40)
```

```python
# Dictionary to store results
energy_saving_results = {}
perf_threshold = 0.95

for app in apps:
    app_path = os.path.join(power_cap_res_path, app)
    
    if not os.path.exists(app_path):
        continue

    max_energy_saving = -float('inf')  # Track max energy saving
    best_power_pair = None

    # Load baseline data (200_250)
    baseline_cpu_file = os.path.join(app_path, "200_250_cpu_power.csv")
    baseline_gpu_file = os.path.join(app_path, "200_250_gpu_metrics.csv")

    if not os.path.exists(baseline_cpu_file) or not os.path.exists(baseline_gpu_file):
        continue

    baseline_cpu_data = pd.read_csv(baseline_cpu_file)
    baseline_gpu_data = pd.read_csv(baseline_gpu_file)

    # Baseline execution time from the last row of Time (s)
    baseline_exec_time = baseline_cpu_data["Time (s)"].iloc[-1]

    # Compute baseline CPU and GPU energy consumption
    baseline_cpu_energy = (baseline_cpu_data["Package Power (W)"].mean() + 
                           baseline_cpu_data["DRAM Power (W)"].mean()) * baseline_exec_time

    baseline_gpu_energy = baseline_gpu_data["Power (W)"].mean() * baseline_exec_time

    # Total baseline energy consumption
    baseline_total_energy = baseline_cpu_energy + baseline_gpu_energy

    for cpu_power, gpu_power in power_cap_pairs:
        power_pair = f"({cpu_power}, {gpu_power})"  # Convert to string format

        # Ensure power pair exists and meets the performance threshold (≥ 0.95)
        if power_pair not in performance_matrix.index or app not in performance_matrix.columns:
            continue

        norm_performance = performance_matrix.loc[power_pair, app]
        if norm_performance < perf_threshold:
            continue  # Skip if performance loss >5%

        # Compute actual execution time
        norm_exec_time = 1 / norm_performance

        # Load actual CPU and GPU power data
        selected_cpu_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_cpu_power.csv")
        selected_gpu_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_gpu_metrics.csv")

        if not os.path.exists(selected_cpu_file) or not os.path.exists(selected_gpu_file):
            continue

        selected_cpu_data = pd.read_csv(selected_cpu_file)
        selected_gpu_data = pd.read_csv(selected_gpu_file)

        # Execution time from the last row of Time (s)
        selected_exec_time = selected_cpu_data["Time (s)"].iloc[-1]

        # Compute CPU and GPU energy consumption
        selected_cpu_energy = (selected_cpu_data["Package Power (W)"].mean() + 
                               selected_cpu_data["DRAM Power (W)"].mean()) * selected_exec_time

        selected_gpu_energy = selected_gpu_data["Power (W)"].mean() * selected_exec_time

        # Total selected energy consumption
        selected_total_energy = selected_cpu_energy + selected_gpu_energy

        # Compute energy savings percentage
        energy_saving = ((baseline_total_energy - selected_total_energy) / baseline_total_energy) * 100

        # Find the power cap pair with the largest energy saving
        if energy_saving > max_energy_saving:
            max_energy_saving = energy_saving
            best_power_pair = (cpu_power, gpu_power)
            best_exec_time = selected_exec_time
            best_energy = selected_total_energy

    # Store results
    if best_power_pair:
        energy_saving_results[app] = {
            "Best Power Pair": f"{best_power_pair[0]}_{best_power_pair[1]}",
            "Baseline Execution Time (s)": baseline_exec_time,
            "Selected Execution Time (s)": best_exec_time,
            "Runtime Overhead (%)": ((best_exec_time - baseline_exec_time) / baseline_exec_time) * 100,
            "Baseline Energy (J)": baseline_total_energy,
            "Selected Energy (J)": best_energy,
            "Energy Saving (%)": max_energy_saving
        }
```

```python
# Print energy saving results
for app, results in energy_saving_results.items():
    print(f"Application: {app}")
    print(f"Best Power Pair: {results['Best Power Pair']}")
    # print(f"Baseline Execution Time (s): {results['Baseline Execution Time (s)']:.4f}")
    # print(f"Selected Execution Time (s): {results['Selected Execution Time (s)']:.4f}")
    print(f"Runtime Overhead (%): {results['Runtime Overhead (%)']:.2f}%")
    # print(f"Baseline Energy (J): {results['Baseline Energy (J)']:.2f}")
    # print(f"Selected Energy (J): {results['Selected Energy (J)']:.2f}")
    print(f"Energy Saving (%): {results['Energy Saving (%)']:.2f}%")
    print("-" * 40)
```

# Energy Saving

```python
import ast
import pandas as pd
import os
from itertools import product

# Define paths
performance_matrix_paths = {
    "altis": "./prediction_res/performance_matrix_gpu_altis_a100.csv",
    "ecp": "./prediction_res/performance_matrix_gpu_ecp_a100.csv",
    "hec": "./prediction_res/performance_matrix_gpu_hec_a100.csv"
    
}
power_cap_paths = {
    "altis": "./altis_power_cap_res/runs/run1/",
    "ecp": "./ecp_power_cap_res/runs/run1/",
    "hec": "./hec_power_cap_res/runs/run1/"
    
}

# Applications mapped to each matrix
apps_by_matrix = {
    "altis": ["sort","pathfinder","lavamd","cfd"],
    "ecp": ['miniGAN', 'sw4lite', 'bert_large', 'UNet', 'Resnet50', 'lammps', 'gromacs', 'NAMD', 'Laghos', 'XSBench'],
    "hec": ["aobench", "background-subtract", "convolution3D",
            "extrema", "knn", "softmax", "zmddft", "zoom", "chacha20", "kalman", "stencil3d","addBiasResidualLayerNorm"]
    
}


# Load performance matrices
performance_matrices = {key: pd.read_csv(path, index_col=0) for key, path in performance_matrix_paths.items()}

# Power cap ranges
cpu_caps = [200, 190, 180, 170, 160, 150, 140]
gpu_caps = [250, 240, 230, 220, 210, 200, 190, 180, 170, 160, 150]
power_cap_pairs = list(product(cpu_caps, gpu_caps))

# Threshold for performance loss
threshold = 0.1

# Store all valid power pairs per app
energy_saving_results = {}

for matrix_key, app_list in apps_by_matrix.items():
    performance_matrix = performance_matrices[matrix_key]
    app_path_base = power_cap_paths[matrix_key]

    for app in app_list:
        app_path = os.path.join(app_path_base, app)

        baseline_pair = "(200, 250)"
        if baseline_pair not in performance_matrix.index or app not in performance_matrix.columns:
            
            continue

        baseline_perf = performance_matrix.loc[baseline_pair, app]

        baseline_cpu_file = os.path.join(app_path, "200_250_cpu_power.csv")
        baseline_gpu_file = os.path.join(app_path, "200_250_gpu_metrics.csv")

        if not os.path.exists(baseline_cpu_file) or not os.path.exists(baseline_gpu_file):
            print(app)
            continue

        baseline_cpu_data = pd.read_csv(baseline_cpu_file)
        baseline_gpu_data = pd.read_csv(baseline_gpu_file)

        baseline_exec_time = baseline_gpu_data["Time (s)"].iloc[-1]
        baseline_cpu_energy = (baseline_cpu_data["Package Power (W)"].mean() +
                               baseline_cpu_data["DRAM Power (W)"].mean()) * baseline_exec_time
        baseline_gpu_energy = baseline_gpu_data["Power (W)"].mean() * baseline_exec_time
        baseline_total_energy = baseline_cpu_energy + baseline_gpu_energy

        valid_power_pairs = []

        for cpu_power, gpu_power in power_cap_pairs:
            pair_key = f"({cpu_power}, {gpu_power})"
            if pair_key not in performance_matrix.index:
                continue
        
            # Predicted normalized performance
            norm_perf = performance_matrix.loc[pair_key, app]
            predicted_perf_loss = 1 - (norm_perf / baseline_perf)
        
            if predicted_perf_loss > threshold:
                continue
        
            cpu_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_cpu_power.csv")
            gpu_file = os.path.join(app_path, f"{cpu_power}_{gpu_power}_gpu_metrics.csv")
        
            if not os.path.exists(cpu_file) or not os.path.exists(gpu_file):
                continue
        
            cpu_data = pd.read_csv(cpu_file)
            gpu_data = pd.read_csv(gpu_file)
        
            # Actual performance from execution time
            actual_exec_time = gpu_data["Time (s)"].iloc[-1]
            actual_perf = baseline_exec_time / actual_exec_time
            actual_perf_loss = 1 - actual_perf
        
            if actual_perf_loss > threshold:
                continue
        
            # Energy computation
            cpu_energy = (cpu_data["Package Power (W)"].mean() + cpu_data["DRAM Power (W)"].mean()) * actual_exec_time
            gpu_energy = gpu_data["Power (W)"].mean() * actual_exec_time
            total_energy = cpu_energy + gpu_energy
        
            energy_saving = baseline_total_energy - total_energy
            energy_saving_percent = energy_saving / baseline_total_energy * 100
        
            # Predicted energy upper bound
            predicted_energy = (cpu_power + gpu_power) * (1 / norm_perf)
            predicted_baseline_energy = (200 + 250) * 1
            predicted_energy_saving_percent = (predicted_baseline_energy - predicted_energy) / predicted_baseline_energy * 100



            baseline_total_power = 200 + 250

            # Average power for the current pair
            avg_cpu_power = cpu_data["Package Power (W)"].mean() + cpu_data["DRAM Power (W)"].mean()
            avg_gpu_power = gpu_data["Power (W)"].mean()
            avg_total_power = avg_cpu_power + avg_gpu_power
            
            # Normalize average power by baseline
            normalized_avg_power = avg_total_power / baseline_total_power
            
            # Energy efficiency: actual performance / normalized power
            energy_efficiency = actual_perf / normalized_avg_power
            
                        

            valid_power_pairs.append({
                "Power Pair": f"{cpu_power}_{gpu_power}",
                "Predicted Performance Loss (%)": predicted_perf_loss * 100,
                "Actual Performance Loss (%)": actual_perf_loss * 100,
                "Energy Saving (%)": energy_saving_percent,
                "Predicted Energy Saving (%)": predicted_energy_saving_percent,
                "Average Power (W)": avg_total_power,
                "Energy Efficiency": energy_efficiency
            })

        # Mark the best predicted energy saving
        if valid_power_pairs:
            df = pd.DataFrame(valid_power_pairs)
            best_idx = df["Predicted Energy Saving (%)"].idxmax()
            df["Selected"] = False
            df.loc[best_idx, "Selected"] = True
            energy_saving_results[app] = df.to_dict(orient="records")


        if valid_power_pairs:
            energy_saving_results[app] = valid_power_pairs
```

```python
for app, results in energy_saving_results.items():
    print(f"\n=== {app} ===")
    df = pd.DataFrame(results)

    # Find the row with the max predicted energy saving
    best_idx = df["Predicted Energy Saving (%)"].idxmax()

    # Format all values to strings with two decimal places for floats
    formatted_rows = []
    for i, row in df.iterrows():
        formatted_row = []
        for val in row.values:
            if isinstance(val, float):
                formatted_row.append(f"{val:.2f}")
            else:
                formatted_row.append(str(val))
        formatted_row.append("✔" if i == best_idx else "")
        formatted_rows.append(formatted_row)

    # Add 'Mark' column
    headers = list(df.columns) + ["Mark"]

    # Calculate column widths
    col_widths = [max(len(row[i]) for row in ([headers] + formatted_rows)) for i in range(len(headers))]

    # Print header
    header_line = "  ".join(header.ljust(col_widths[i]) for i, header in enumerate(headers))
    print(header_line)

    # Print rows
    for row in formatted_rows:
        row_line = "  ".join(val.ljust(col_widths[i]) for i, val in enumerate(row))
        print(row_line)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

apps = []
actual_energy_saving = []
actual_perf_loss = []

for app, results in energy_saving_results.items():
    df = pd.DataFrame(results)
    if "Predicted Energy Saving (%)" not in df.columns:
        continue

    # Select the row with the highest predicted energy saving
    best_idx = df["Predicted Energy Saving (%)"].idxmax()
    row = df.loc[best_idx]

    apps.append(app)
    actual_energy_saving.append(row["Energy Saving (%)"])

    # Apply the condition to actual performance loss
    perf_loss = row["Actual Performance Loss (%)"]
    if perf_loss < 0 and perf_loss > -3:
        perf_loss = 0.1
    actual_perf_loss.append(perf_loss)

# Plot actual energy savings
plt.figure(figsize=(10, 4))
plt.bar(apps, actual_energy_saving, color='#c99b38')
plt.ylabel("Actual Energy Saving (%)")
plt.ylim(0, 22)
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()

# Plot actual performance loss
plt.figure(figsize=(10, 4))
plt.bar(apps, actual_perf_loss, color='#c99b38')
plt.ylabel("Actual Performance Loss (%)")
plt.ylim(0, 22)
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
# Define paths
performance_matrix_paths = {
    "npb": "./prediction_res/performance_matrix_cpu_a100.csv",
    
}
power_cap_paths = {
    "npb": "./npb_power_cap_res/runs/run2/",
    
}

# Applications mapped to each matrix
apps_by_matrix = {
    "npb": ["ep","ft","lu","LULESH","mg","Nekbone","sp","ua"]
    # "npb": ["LULESH"]
}


# Load performance matrices
performance_matrices = {key: pd.read_csv(path, index_col=0) for key, path in performance_matrix_paths.items()}

# Power cap ranges
cpu_caps = [540,520,500,480,460,440,420,400,380,360,340,320,300,280,260]
# cpu_caps = [520]

# Threshold for performance loss
threshold = 0.1

# Store all valid power pairs per app
energy_saving_results_CPU = {}

for matrix_key, app_list in apps_by_matrix.items():
    performance_matrix = performance_matrices[matrix_key]
    app_path_base = power_cap_paths[matrix_key]

    for app in app_list:
       
        app_path = os.path.join(app_path_base, app)

        baseline_pair = 540
        if baseline_pair not in performance_matrix.index or app not in performance_matrix.columns:
          
            continue

        baseline_perf = performance_matrix.loc[baseline_pair, app]

        baseline_cpu_file = os.path.join(app_path, "540_250_cpu_power.csv")
        baseline_gpu_file = os.path.join(app_path, "540_250_gpu_metrics.csv")

        if not os.path.exists(baseline_cpu_file) or not os.path.exists(baseline_gpu_file):
        
            continue

        baseline_cpu_data = pd.read_csv(baseline_cpu_file)
        baseline_gpu_data = pd.read_csv(baseline_gpu_file)

        baseline_exec_time = baseline_gpu_data["Time (s)"].iloc[-1]
        baseline_cpu_energy = (baseline_cpu_data["Package Power (W)"].mean() +
                               baseline_cpu_data["DRAM Power (W)"].mean()) * baseline_exec_time
        # baseline_gpu_energy = baseline_gpu_data["Power (W)"].mean() * baseline_exec_time
        baseline_total_energy = baseline_cpu_energy 

        valid_power_pairs = []

        for cpu_power in cpu_caps:
            # Predicted normalized performance
            norm_perf = performance_matrix.loc[cpu_power, app]
            predicted_perf_loss = 1 - (norm_perf / baseline_perf)
            # print(predicted_perf_loss)
            if predicted_perf_loss > threshold:
    
                continue
        
            cpu_file = os.path.join(app_path, f"{cpu_power}_{250}_cpu_power.csv")
            gpu_file = os.path.join(app_path, f"{cpu_power}_{250}_gpu_metrics.csv")
        
            if not os.path.exists(cpu_file) or not os.path.exists(gpu_file):
                continue
        
            cpu_data = pd.read_csv(cpu_file)
            gpu_data = pd.read_csv(gpu_file)
        
            # Actual performance from execution time
            actual_exec_time = gpu_data["Time (s)"].iloc[-1]
            actual_perf = baseline_exec_time / actual_exec_time
            actual_perf_loss = 1 - actual_perf
            # print(baseline_exec_time)
            if actual_perf_loss > threshold:
                continue
        
            # Energy computation
            cpu_energy = (cpu_data["Package Power (W)"].mean() + cpu_data["DRAM Power (W)"].mean()) * actual_exec_time
            # gpu_energy = gpu_data["Power (W)"].mean() * actual_exec_time
            total_energy = cpu_energy
        
            energy_saving = baseline_total_energy - total_energy
            energy_saving_percent = energy_saving / baseline_total_energy * 100
        
            # Predicted energy upper bound
            predicted_energy = (cpu_power) * (1 / norm_perf)
            predicted_baseline_energy = (200 + 250) * 1
            predicted_energy_saving_percent = (predicted_baseline_energy - predicted_energy) / predicted_baseline_energy * 100



            baseline_total_power = 540

            # Average power for the current pair
            avg_cpu_power = cpu_data["Package Power (W)"].mean() + cpu_data["DRAM Power (W)"].mean()
            # avg_gpu_power = gpu_data["Power (W)"].mean()
            avg_total_power = avg_cpu_power             
            # Normalize average power by baseline
            normalized_avg_power = avg_total_power / baseline_total_power
            
            # Energy efficiency: actual performance / normalized power
            energy_efficiency = actual_perf / normalized_avg_power
            
                        

            valid_power_pairs.append({
                "Power Pair": f"{cpu_power}_{gpu_power}",
                "Predicted Performance Loss (%)": predicted_perf_loss * 100,
                "Actual Performance Loss (%)": actual_perf_loss * 100,
                "Energy Saving (%)": energy_saving_percent,
                "Predicted Energy Saving (%)": predicted_energy_saving_percent,
                "Average Power (W)": avg_total_power,
                "Energy Efficiency": energy_efficiency
            })

        # Mark the best predicted energy saving
        if valid_power_pairs:
            df = pd.DataFrame(valid_power_pairs)
            best_idx = df["Predicted Energy Saving (%)"].idxmax()
            df["Selected"] = False
            df.loc[best_idx, "Selected"] = True
            energy_saving_results_noCPUCap[app] = df.to_dict(orient="records")


        if valid_power_pairs:
            energy_saving_results_CPU[app] = valid_power_pairs
```

# Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

apps_gpu = []
baseline_efficiency_gpu = []
selected_efficiency_gpu = []
gpu_only_efficiency_gpu = []
baseline_perf_gpu = []
selected_perf_gpu = []
gpu_only_perf_gpu = []

apps_cpu = []
baseline_efficiency_cpu = []
selected_efficiency_cpu = []
baseline_perf_cpu = []
selected_perf_cpu = []

# GPU-enabled applications
for app in energy_saving_results:
    if app not in energy_saving_results_noCPUCap:
        continue

    df = pd.DataFrame(energy_saving_results[app])
    df_gpu = pd.DataFrame(energy_saving_results_noCPUCap[app])

    if "Predicted Energy Saving (%)" not in df.columns or \
       "Predicted Energy Saving (%)" not in df_gpu.columns:
        continue

    best_idx = df["Predicted Energy Saving (%)"].idxmax()
    best_idx_gpu = df_gpu["Predicted Energy Saving (%)"].idxmax()

    row = df.loc[best_idx]
    row_gpu = df_gpu.loc[best_idx_gpu]

    apps_gpu.append(app)

    baseline_efficiency_gpu.append(1.0)
    selected_efficiency_gpu.append(row.get("Energy Efficiency", 1.0))
    gpu_only_efficiency_gpu.append(row_gpu.get("Energy Efficiency", 1.0))

    baseline_perf_gpu.append(1.0)
    selected_perf_gpu.append(1 - row.get("Actual Performance Loss (%)", 0) / 100)
    gpu_only_perf_gpu.append(1 - row_gpu.get("Actual Performance Loss (%)", 0) / 100)

# CPU-only applications
for app in energy_saving_results_CPU:
    df = pd.DataFrame(energy_saving_results_CPU[app])
    if "Predicted Energy Saving (%)" not in df.columns:
        continue

    best_idx = df["Predicted Energy Saving (%)"].idxmax()
    row = df.loc[best_idx]

    apps_cpu.append(app)

    baseline_efficiency_cpu.append(1.0)
    selected_efficiency_cpu.append(row.get("Energy Efficiency", 1.0))

    baseline_perf_cpu.append(1.0)
    selected_perf_cpu.append(1 - row.get("Actual Performance Loss (%)", 0) / 100)

# Plot setup
x = np.arange(len(apps_gpu))
width = 0.25

app_names = [
    "sort","pathfinder","lavamd","cfd",
    'miniGAN', 'sw4lite', 'bert_large', 'UNet', 'Resnet50', 'lammps', 'gromacs', 'NAMD', 'Laghos', 'XSBench',
    "aobench", "subtract", "convolution3D", "extrema", "knn", "softmax", "zmddft", "zoom", "chacha20", "kalman", "stencil3d","residualLayerNorm"
]
x = np.arange(len(app_names))
```

```python
fig, ax = plt.subplots(figsize=(9, 2.5))

# Plot bars with edge outline
ax.bar(x - width, baseline_efficiency_gpu, width, label="No Power Cap", color="#001F3F", edgecolor='black')
ax.bar(x, gpu_only_efficiency_gpu, width, label="GPU Cap Only", color="#5DA9B8", edgecolor='black')
ax.bar(x + width, selected_efficiency_gpu, width, label=r"OPEN ($\gamma$=10%)", color="#C6452C", hatch='/', edgecolor='black')

# Axes and labels
ax.set_ylim(0.6, 2.5)
ax.set_ylabel("Energy Efficiency", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(app_names, rotation=30)
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=15)

# Legend and grid
ax.legend(ncol=3, loc="upper center", bbox_to_anchor=(0.5, 1.1))
ax.grid(linewidth=0.5)

# Layout and save
plt.tight_layout()
plt.savefig("./fig/energy_efficiency.png", dpi=300, bbox_inches='tight')
plt.show()
```
